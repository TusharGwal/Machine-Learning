## Importing the libraries

````
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

## Importing the dataset

````
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
````

## Splitting the dataset into the Training set and Test set

````
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
````

## Feature Scaling

````
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
````

## Training the SVM model on the Training set

````
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
````

You're getting deep into **kernel selection** in Support Vector Machines (SVMs) — and understanding **why we use a `'linear'` kernel vs an `'rbf'` (Radial Basis Function)** is key to mastering SVMs.

Let’s break it down clearly:

---

## 🧠 First, What is a Kernel?

> A **kernel** is a trick SVM uses to **map the data into a higher-dimensional space** so it can find a separating hyperplane more easily — especially when data is **not linearly separable**.

---

## ✅ `'linear'` Kernel — Use When Data Is Linearly Separable

```python
SVC(kernel='linear')
```

### 🎯 When to use it:

* Your data **can be separated with a straight line or flat hyperplane**
* The relationship between input features and the target is **linear**
* You care about **speed**, especially for large datasets

✅ Pros:

* Fast training time
* Simple and interpretable
* Good when the number of features is **high** (like in text classification, TF-IDF, etc.)

---

## 🌈 `'rbf'` (Radial Basis Function) Kernel — Use When Data Is Nonlinear

```python
SVC(kernel='rbf')
```

### 🎯 When to use it:

* Your data is **not linearly separable**
* You suspect that the decision boundary is **curved** or complex
* You want the model to capture **nonlinear relationships**

✅ Pros:

* Powerful and flexible
* Works well when there’s no clear linear boundary

⚠️ Cons:

* More **computationally expensive**
* **Harder to interpret**

---

## 🔍 Visual Example

### Linear kernel:

```
 ● ● ● | ○ ○ ○
```

A straight line separates the classes.

### RBF kernel:

```
        ○ ○
      ○     ○
    ●         ●
      ●     ●
        ● ●

```

No straight line can separate — needs a **curved boundary** → use RBF.

---

## 🤔 So Why Use `kernel='linear'`?

Because:

* If you know your dataset is **linearly separable**, or nearly so
* Or you're working with **high-dimensional but sparse** features (e.g., text, bag-of-words)
* Or you just want a **faster, interpretable baseline**

Then `kernel='linear'` is the best place to start.

---

## 🧪 How to Decide in Practice?

### ✅ Try both kernels and compare accuracy:

```python
from sklearn.svm import SVC

linear_clf = SVC(kernel='linear').fit(X_train, y_train)
rbf_clf = SVC(kernel='rbf').fit(X_train, y_train)

print("Linear Accuracy:", linear_clf.score(X_test, y_test))
print("RBF Accuracy:", rbf_clf.score(X_test, y_test))
```

---

## 📌 Summary

| Kernel     | When to Use                         | Pros                | Cons                            |
| ---------- | ----------------------------------- | ------------------- | ------------------------------- |
| `'linear'` | Data is (almost) linearly separable | Fast, interpretable | Can't handle nonlinear patterns |
| `'rbf'`    | Data is nonlinear                   | Flexible, accurate  | Slower, less interpretable      |

## Predicting the Test set results

## Visualising the Training set results

````
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
````
